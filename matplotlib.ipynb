```python
from matplotlib import pyplot as plt

x = [5,8,10]
y = [12,16,6]

plt.plot(x,y)

plt.title("Information")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True,color='k')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x1=[5,10,15]
y1=[20,30,40]
x2=[20,30,40]
y2=[5,10,14]

plt.plot(x1,y1,'g',label='line one',linewidth=2)
plt.plot(x2,y2,'b',label='line two',linewidth=2)

plt.title("Epic")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")

plt.legend()

plt.grid(True,color='k')

plt.show()
```

# Bargraph

```python
from matplotlib import pyplot as plt

plt.bar([1,3,5,7,9],[5,2,7,8,2],label="Example One")
plt.bar([2,4,6,8,10],[8,6,2,5,6],label="Example Two",color='g')

plt.legend()

plt.title("Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

# Histogram

```python
from matplotlib import pyplot as plt

ages=[22,25,30,40,50,60,70,80,90,100,120,150]
bins=[0,1,2,4,10,20,30,50,60,70,80,90]

plt.hist(ages,bins,histtype='bar',rwidth=0.8)

plt.title("Histogram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

# Scatter plot

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]

plt.scatter(x,y,label='skitscat',color='k')

plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

# Stack Plot

```python
from matplotlib import pyplot as plt

days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],color='m',label='sleeping',linewidth=5)
plt.plot([],color='c',label='eating',linewidth=5)
plt.plot([],color='r',label='working',linewidth=5)
plt.plot([],color='k',label='playing',linewidth=5)

plt.stack(days,sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.title("Stack Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.legend
```
